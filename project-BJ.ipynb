INteractive scatter plot comparing 3PA and 3p%

```python
import plotly.express as px
import pandas as pd
import statsmodels.api as sm

# Convert necessary columns
nba_df['3PA'] = pd.to_numeric(nba_df['3PA'], errors='coerce')
nba_df['3P%'] = pd.to_numeric(nba_df['3P%'], errors='coerce')
nba_df.dropna(subset=['3PA', '3P%'], inplace=True)

# Fit OLS regression using statsmodels
X = sm.add_constant(nba_df['3PA'])  # Add intercept
model = sm.OLS(nba_df['3P%'], X).fit()
nba_df['Predicted_3P%'] = model.predict(X)

# Create interactive plot
fig = px.scatter(
    nba_df,
    x='3PA',
    y='3P%',
    color='Season',
    hover_name='Team',
    hover_data={'Season': True, '3PA': True, '3P%': True},
    title='3-Point Attempts vs. 3-Point Percentage (Interactive)',
    template='plotly_white'
)

# Add regression line
fig.add_traces(
    px.line(
        nba_df.sort_values('3PA'),
        x='3PA',
        y='Predicted_3P%',
        labels={'Predicted_3P%': 'Trend Line'}
    ).data
)

# Final layout tweaks
fig.update_layout(
    xaxis_title='3-Point Attempts (3PA)',
    yaxis_title='3-Point Percentage (3P%)',
    height=600,
    showlegend=False
)

fig.show()
```

```python
csv_folder = os.path.expanduser(fr"C:\Users\shank\OneDrive\Desktop\Final-Project-4\Resources")

combined_data = []

for year in range(2014, 2026):
    # Load the team and opponent data
    team_path = os.path.join(csv_folder, f"Team_{year}.csv")
    opp_path = os.path.join(csv_folder, f"opponents_{year}.csv")

    team_df = pd.read_csv(team_path)
    opp_df = pd.read_csv(opp_path)

    # Add season
    team_df['Season'] = year
    opp_df['Season'] = year

    # Prefix all columns in opponent dataframe (except 'Team' and 'Season')
    opp_df = opp_df.rename(columns={
        col: f"o{col}" for col in opp_df.columns if col not in ["Team", "Season"]
    })

    # Merge on Team and Season
    merged_df = pd.merge(team_df, opp_df, on=["Team", "Season"], how="inner")
    combined_data.append(merged_df)

# Combine all years into one DataFrame
nba_df_full = pd.concat(combined_data, ignore_index=True)
nba_df_full.head(60)
```

```python
def probability_to_moneyline(p):
    epsilon = 1e-6  # Small number to prevent division by zero
    p = max(min(p, 1 - epsilon), epsilon)  # Clamp between epsilon and 1 - epsilon
    if p >= 0.5:
        return -round((p / (1 - p)) * 100)
    else:
        return round(((1 - p) / p) * 100)

# Create a new DataFrame to hold predictions only for those rows
nba_filtered = nba_df_full.loc[X.index].copy()

# Add predictions
nba_filtered['Win_Probability'] = model.predict_proba(X)[:, 1]
nba_filtered['Moneyline'] = nba_filtered['Win_Probability'].apply(probability_to_moneyline)
```
