```python
import pandas as pd
import os

# Set the path to your resources folder
csv_folder = os.path.expanduser("~/Desktop/Final-Project-4/Resources")

# List to collect DataFrames
all_years = []

# Loop through 2014 to 2025 and read each CSV
for year in range(2014, 2026):
    file_name = f'Team_{year}.csv'
    file_path = os.path.join(csv_folder, file_name)

    # Read and label each file with the season
    df = pd.read_csv(file_path)
    df['Season'] = year
    all_years.append(df)

# Combine all into one DataFrame
nba_df = pd.concat(all_years, ignore_index=True)

# Quick preview
print(nba_df.shape)
nba_df.head()
```
