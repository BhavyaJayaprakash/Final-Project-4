```python
csv_folder = os.path.expanduser("~/Desktop/Final-Project-4/Resources")

combined_data = []

for year in range(2014, 2026):
    # Load the team and opponent data
    team_path = os.path.join(csv_folder, f"Team_{year}.csv")
    opp_path = os.path.join(csv_folder, f"opponents_{year}.csv")

    team_df = pd.read_csv(team_path)
    opp_df = pd.read_csv(opp_path)

    # Add season
    team_df['Season'] = year
    opp_df['Season'] = year

    # Prefix all columns in opponent dataframe (except 'Team' and 'Season')
    opp_df = opp_df.rename(columns={
        col: f"o{col}" for col in opp_df.columns if col not in ["Team", "Season"]
    })

    # Merge on Team and Season
    merged_df = pd.merge(team_df, opp_df, on=["Team", "Season"], how="inner")
    combined_data.append(merged_df)

# Combine all years into one DataFrame
nba_df_full = pd.concat(combined_data, ignore_index=True)
nba_df_full.head(60)
```

```python
nba_df_full.columns
```

```python
# Training = 2014–2024
train_df = nba_df_full[nba_df_full['Season'] < 2025].copy()

# Testing = 2025
test_df = nba_df_full[nba_df_full['Season'] == 2025].copy()
```

```python
champions = {
    2014: 'San Antonio Spurs',
    2015: 'Golden State Warriors',
    2016: 'Cleveland Cavaliers',
    2017: 'Golden State Warriors',
    2018: 'Golden State Warriors',
    2019: 'Toronto Raptors',
    2020: 'Los Angeles Lakers',
    2021: 'Milwaukee Bucks',
    2022: 'Golden State Warriors',
    2023: 'Denver Nuggets',
    2024: 'Boston Celtics'
}

train_df['Champion'] = train_df.apply(
    lambda row: 1 if champions.get(row['Season'], '') in row['Team'] else 0,
    axis=1
)
```

```python
#Features included
features = ['FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA',
            '2P%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST',
            'STL', 'TOV', 'PF','PTS', 'o3P%', 'o2P%', 'oPTS']

#Training
X = train_df[features]
y = train_df['Champion'] 

#Get the test data features
X_test = test_df[features]
```

```python
from sklearn.ensemble import RandomForestClassifier

#Model
model = RandomForestClassifier(random_state=72)

#Train the model on historical data
model.fit(X, y)
```

```python
#Predict the probability of being the champion for 2025 teams
y_probs = model.predict_proba(X_test)[:, 1]

#Test set prediction
test_df['Champion_Prob'] = y_probs

#Sost likely 2025 NBA Champion
teams_prob = test_df.sort_values(by='Champion_Prob', ascending=False)

teams_prob[['Team', 'Champion_Prob']].head(30)
```
